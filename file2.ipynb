```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('./dataset/tmdb_5000_movies.csv')
credit = pd.read_csv('./dataset/tmdb_5000_credits.csv')
```

```python
movie.head(1)
```

```python
movies = movie.merge(credit, on='title')
```

```python
movie.shape
```

```python
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    if isinstance(obj, list):
        return obj
    
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convertCast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def get_Director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_Director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

New dataframe with tags column as string

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

joined list by spaces

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df
```

Lower case

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df
```

```python
new_df['tags'][0]
```

Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()  #Stemming

```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

###Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words= 'english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()   #vectorization  
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
print(cv.get_feature_names_out())
```

Finding Cosine Similarity between each movie

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie_title):
    movie_index = new_df[new_df['title'] == movie_title].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse = True , key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman")
```
